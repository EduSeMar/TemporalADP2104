```python
%matplotlib inline
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../../Datasets/melbourne_housing-clean.csv', index_col=0)

df.head()
```

Vamos a crear un histograma con los estilos default:

```python
# El histograma con la instrucci칩n viejita de la variable 'price'

sns.distplot(df['price'], kde=False, norm_hist=False);
```

```python
# La versi칩n m치s sencilla

sns.histplot(df['price'], kde=False)
```

Seaborn y matplotib incluyen algunos estilos predefinidos que podemos usar para cambiar el estilo visual de nuestras gr치ficas:

```python
# El estilo 'white'

sns.set_style('white')

sns.histplot(df['price'], kde=False)
```

```python
sns.set_style('whitegrid')

sns.histplot(df['price'], kde=False);
```

```python
sns.set_style('darkgrid')

sns.histplot(df['price'], kde=False)
```

```python
plt.style.use('fivethirtyeight')

sns.histplot(df['price'], kde=False);
```

```python
plt.style.use('ggplot') ## ggplot es una librer칤a muy populat
# de graficaci칩n en R

sns.histplot(df['price'], kde=False)
```

```python
plt.style.use('classic')

sns.histplot(df['price'], kde=False)
```

Tambi칠n podemos usar estilos de manera 'temporal'. Esto quiere decir que el estilo s칩lo aplica a las gr치ficas que realicemos dentro de la sentencia `with`:

```python
plt.style.use('default')

with sns.axes_style('ticks'):
    sns.histplot(df['price'], kde=False)
```

```python
with plt.style.context('seaborn-bright'):
    sns.histplot(df['price'], kde=False)
```
