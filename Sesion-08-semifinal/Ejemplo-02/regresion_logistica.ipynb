Tenemos el siguiente dataset con mediciones tomadas a diversos tumores de mama, junto con el diagnóstico que fue hecho. El diagnóstico puede ser 'M' para maligno y 'B' para benigno:

El algoritmo de regresión logística es método supervisado que sirve para hacer clasificaciones. Se tiene una variable respuesta $y$ que toma dos categoría (0/1), (Verdadero/Falso), (Si/No) y si utilizan k variables $X_1,X_2,...,X_k$ para determinar si se es del tipo 0 o del tipo 1.

```python
import pandas as pd
```

```python
df = pd.read_csv('../../Datasets/breast_cancer_wisconsin.csv', index_col=0)

df.head()
```

Primero que nada vamos a limpiar nuestro dataset un poco:

```python
# Quito la última columna, pues no tiene sentido

df = df.drop(columns=['Unnamed: 32'])
```

```python
df.head()
```

```python
# Nuestra respuesta será la variable 'diagnosis'

df['diagnosis'].unique()
```

```python
# El algoritmo requiere que esta variable este codificada en 0/1

df['diagnosis'] = df['diagnosis'].apply(lambda x: 1 if x == 'M' else 0)
```

```python
df.head()
```

Vamos a intentar entrenar un modelo de Regresión Logística para clasificar tumores como malignos o benignos.

```python
# Con base en el resto de las columnas vamos a intentar determinar si el tumor es
# benigno (0) o maligno (1)

X = df.drop(columns=['diagnosis']) # todas la columnas expecto 'diagnosis'
y = df['diagnosis'] # la columna 'diagnosis' será variable respuesta
```

```python
# La sublibrería que ocuparemos es LogisticRegression de scikit learn

from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
# Dividimos nuestros datos en entrenamiento y prueba

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
X_train
```

```python
# DEfinimos un objeto tipo Regresión Logística
logreg = LogisticRegression()
```

```python
# Ajustamos con nuestros datos de entrenamiento
logreg.fit(X_train, y_train)
```

Ya que tenemos un modelo, podemos hacer predicciones utilizando el dataset de prueba:

```python
# Ocupamos este modelo para hacer predicciones basándonos en nuestro set de prueba
# esto lo hacemos con el metodo predict

y_pred = logreg.predict(X_test)

y_pred
```

¿Cómo sabemos si esto es correcto o no? Bueno, LogisticRegression tiene un método `score` para obtener la exactitud del modelo:

```python
# se obtiene con el método score

logreg.score(X_test, y_test)
```

El score de este modelo se llama `mean accuracy` (exactitud promedio). Este valor puede encontrarse en un intervalo entre 0 y 1, con 1 siendo 'exactitud perfecta'.
