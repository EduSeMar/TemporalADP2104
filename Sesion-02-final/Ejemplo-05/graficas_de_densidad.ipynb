Podemos obtener gráficas de densidad usando el método `kdeplot` de Seaborn, sólo cambiando los argumentos que le pasamos:

```python
import numpy as np
import seaborn as sns
import pandas as pd

sns.set(style='whitegrid')
```

```python
laplace = np.random.laplace(loc=0.0, scale=1, size=10000) #genero 10,000 números aletorios
# de la distribución Laplace

# La función que me ayuda a dibujar densidades es la función kdeplot()
sns.kdeplot(laplace);
```

Observemos que acá se está graficando frecuancias relativas, i.e. porcentajes con respecto al todo, es por eso que el punto más alto de esta densidad es aprox 0.43

```python
# Otro ejemplo
# ahora simularé de otra distrubución, la ji.cuadrada, pero no se preocupen
# estoy simulando datos para obtener rapidamente datos

datos = np.random.chisquare(4, size=10000)
sns.kdeplot(datos)
```

Las gráficas de densidad son muy útiles para graficar varias distribuciones en la misma gráfica y poderlas comparar:

```python
sns.set(style='white')

#Genero datos aleatorios
primer_cjto = np.random.normal(loc=-2, scale=3, size=10000)
segundo_cjto = np.random.normal(loc=4.5, scale=1, size=10000)
tercer_cjto = np.random.exponential(scale=1.0, size=10000) - 1

sns.kdeplot(primer_cjto,shade=True)
sns.kdeplot(segundo_cjto,shade=True)
sns.kdeplot(tercer_cjto,shade=True)
```

Como puedes ver, usando gráficas de densidad obtenemos mucha claridad en la comparación. En este ejemplo agregamos la bandera `shade=True` para obtener el color de relleno de cada distribución. En una sesión posterior aprenderemos cómo estilizar nuestras gráficas a profundidad.

```python
df = pd.read_csv('../../Datasets/melbourne_housing-clean.csv')
```

```python
sns.kdeplot(df['price'],shade=True)
```
